```python
"""
Exercise 1: Assign documents to 2 clusters using K-Means + TF-IDF
"""

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

documents = [
    "The plaintiff filed a lawsuit for breach of contract.",
    "Contract disputes are common in civil litigation.",
    "The defendant denies all allegations of negligence.",
    "Personal injury and negligence claims require proof of harm."
]

# Step 1: Convert documents to TF-IDF vectors
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

# Step 2: Apply K-Means with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(X)

# Step 3: Display results
print("=" * 55)
print("        Document Clustering Results (K=2)        ")
print("=" * 55)
for i, (doc, label) in enumerate(zip(documents, kmeans.labels_)):
    print(f"\nDocument {i + 1} → Cluster {label + 1}")
    print(f"  \"{doc}\"")

print("\n" + "=" * 55)
print("Cluster Summary:")
for cluster_id in range(2):
    print(f"\n  Cluster {cluster_id + 1}:")
    for i, label in enumerate(kmeans.labels_):
        if label == cluster_id:
            print(f"    - Document {i + 1}")
```

```python
"""
Exercise 2: Measure how close documents are using Cosine Similarity + TF-IDF
"""

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

doc1 = "The tenant failed to pay rent on time, violating the lease agreement."
doc2 = "The lease was breached because the monthly rent payment was late."
doc3 = "The company merged with a competitor to dominate the market."

documents = [doc1, doc2, doc3]

# Step 1: Vectorize using TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# Step 2: Compute cosine similarity
sim_1_2 = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]
sim_1_3 = cosine_similarity(tfidf_matrix[0], tfidf_matrix[2])[0][0]

# Step 3: Display results
print("=" * 55)
print("       Text Similarity Results (TF-IDF + Cosine)     ")
print("=" * 55)
print(f"\nDoc1: \"{doc1}\"")
print(f"\nDoc2: \"{doc2}\"")
print(f"\nDoc3: \"{doc3}\"")

print("\n" + "-" * 55)
print(f"  Similarity(Doc1, Doc2): {sim_1_2:.4f}  →  ", end="")
print("High similarity ✅" if sim_1_2 > 0.4 else "Low similarity ❌")

print(f"  Similarity(Doc1, Doc3): {sim_1_3:.4f}  →  ", end="")
print("High similarity ✅" if sim_1_3 > 0.4 else "Low similarity ❌")

print("\n" + "=" * 55)
print("Interpretation:")
print("  Score close to 1.0 = Very similar")
print("  Score close to 0.0 = Very different")
```

```python
"""
Exercise 3: Measure text similarity using SentenceTransformer (LLM embeddings)
"""

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Documents
doc1 = "The tenant failed to pay rent on time, violating the lease agreement."
doc2 = "The lease was breached because the monthly rent payment was late and not sent to the appropriate email."
doc3 = "The company merged with a competitor to dominate the market."

# Step 1: Load pre-trained SentenceTransformer model
print("Loading SentenceTransformer model...")
model = SentenceTransformer("all-MiniLM-L6-v2")  # Lightweight, fast, and accurate

# Step 2: Encode documents into embeddings
print("Encoding documents...\n")
embeddings = model.encode([doc1, doc2, doc3])

# Step 3: Compute cosine similarity
sim_1_2 = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]
sim_1_3 = cosine_similarity([embeddings[0]], [embeddings[2]])[0][0]

# Step 4: Display results
print("=" * 60)
print("   Text Similarity Results (SentenceTransformer + Cosine)  ")
print("=" * 60)
print(f"\nDoc1: \"{doc1}\"")
print(f"\nDoc2: \"{doc2}\"")
print(f"\nDoc3: \"{doc3}\"")

print("\n" + "-" * 60)
print(f"  Similarity(Doc1, Doc2): {sim_1_2:.4f}", end="  →  ")
if sim_1_2 > 0.7:
    print("Very High Similarity ✅✅")
elif sim_1_2 > 0.4:
    print("Moderate Similarity ✅")
else:
    print("Low Similarity ❌")

print(f"  Similarity(Doc1, Doc3): {sim_1_3:.4f}", end="  →  ")
if sim_1_3 > 0.7:
    print("Very High Similarity ✅✅")
elif sim_1_3 > 0.4:
    print("Moderate Similarity ✅")
else:
    print("Low Similarity ❌")

print("\n" + "=" * 60)
print("Model used: all-MiniLM-L6-v2 (SentenceTransformer)")
print("Metric    : Cosine Similarity (range: 0.0 to 1.0)")
print("=" * 60)
```
