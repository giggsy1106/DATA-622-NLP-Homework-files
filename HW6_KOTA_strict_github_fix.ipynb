```python
import random
import re
import os

def load_text(path):
    with open(path, "r", encoding="utf-8") as f:
        return f.read()

def build_markov_chain(words):
    chain = {}
    for i in range(len(words) - 1):
        curr_word = words[i]
        next_word = words[i + 1]
        chain.setdefault(curr_word, []).append(next_word)
    return chain

def generate_story(chain, start_phrase, length=150):
    words = start_phrase.lower().split()
    current = words[-1]

    for _ in range(length):
        if current not in chain:
            # If the current word is not in the chain, pick a random word to continue
            current = random.choice(list(chain.keys()))
        # Handle cases where the current word might not have any next words (though unlikely with a good dataset)
        if not chain.get(current):
            current = random.choice(list(chain.keys()))
        next_word = random.choice(chain[current])
        words.append(next_word)
        current = next_word

    story = " ".join(words)
    return story.capitalize() + "."

# The TRAINING variable from the kernel state contains the text to be used.
# The content is truncated for display here but the full content from the kernel state will be used.
TRAINING = "\nOnce upon a time there was a kingdom far away in the mountains. The kingdom was ruled by a wise and kind king.\nThe king had a daughter who was brave and clever. One day a dragon appeared near the kingdom and frightened the villagers.\nThe brave princess decided to face the dragon alone. She walked through the dark forest and found the dragon sleeping near a cave.\nThe princess spoke gently to the dragon and discovered it was lonely and lost. She invited the dra"

def main():
    data_dir = "data"
    data_path = os.path.join(data_dir, "fairy_tale.txt")

    # Ensure the 'data' directory exists
    os.makedirs(data_dir, exist_ok=True)

    # Write the TRAINING data to the fairy_tale.txt file
    # This ensures the file exists when load_text tries to read it
    with open(data_path, "w", encoding="utf-8") as f:
        f.write(TRAINING)

    text = load_text(data_path)

    words = re.findall(r"\b\w+\b", text.lower())
    chain = build_markov_chain(words)

    start = "Once upon a time there was a kingdom"
    story = generate_story(chain, start)

    print("\n--- GENERATED MARKOV STORY ---\n")
    print(story)

if __name__ == "__main__":
    main()
```

```python
from transformers import pipeline, set_seed

def main():
    generator = pipeline("text-generation", model="gpt2")
    set_seed(42)

    prompt = (
        "Write a creative story explaining why it is important to study data science today. "
        "Make it imaginative, narrative, and engaging."
    )

    output = generator(prompt, max_length=250, num_return_sequences=1)[0]["generated_text"]

    print("\n--- LLM-GENERATED STORY ---\n")
    print(output)

if __name__ == "__main__":
    main()
```
